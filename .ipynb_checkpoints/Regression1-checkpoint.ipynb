```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import pickle
import warnings
warnings.filterwarnings("ignore", category=UserWarning)

# Ensure models folder
os.makedirs("models", exist_ok=True)

# Load Data
df = pd.read_csv("cleaned_data.csv")

# UPDATED FEATURE SETS (REMOVED Total Circulation + Operating Expenditures FROM INPUTS)
# We KEEP Total Circulation and Total Operating Income as TARGETS

features_shared = [
    'Population of Service Area',
    'Total Library Visits',
    'Total Programs (Synchronous + Prerecorded)',
    'Total Program Attendance & Views',
    'Total Collection',
    'Wages & Salaries Expenditures',
    'Library Materials Expenditures',
    'Town Tax Appropriation for Library'
]

# For app compatibility, if you still want full list somewhere, keep features_all but we won't use opEx/totalcirc as inputs:
features_all = [
    'Population of Service Area',
    'Total Library Visits',
    'Total Circulation',
    'Total Programs (Synchronous + Prerecorded)',
    'Total Program Attendance & Views',
    'Total Collection',
    'Operating Expenditures',
    'Wages & Salaries Expenditures',
    'Library Materials Expenditures',
    'Town Tax Appropriation for Library'
]

# PART 1: PREDICTING TOTAL CIRCULATION (TARGET)
print("PREDICTING: Total Circulation")


# Target and inputs (note: Operating Expenditures removed from inputs)
X_circ = df[features_shared].copy()  # inputs for circulation model
y_circ = df['Total Circulation'].copy()  # target remains in dataset

# Split data
X_train_circ, X_test_circ, y_train_circ, y_test_circ = train_test_split(
    X_circ, y_circ, test_size=0.2, random_state=42
)

# Scale the features for circulation
scaler_circ = StandardScaler()
X_train_circ_scaled = scaler_circ.fit_transform(X_train_circ)
X_test_circ_scaled = scaler_circ.transform(X_test_circ)

# Model 1: Linear Regression

lr_circ = LinearRegression()
lr_circ.fit(X_train_circ_scaled, y_train_circ)
pred_circ_lr = lr_circ.predict(X_test_circ_scaled)

# Model 2: Polynomial Regression

poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train_circ_scaled)
X_test_poly = poly.transform(X_test_circ_scaled)
poly_circ = LinearRegression()
poly_circ.fit(X_train_poly, y_train_circ)
pred_circ_poly = poly_circ.predict(X_test_poly)

# Model 3: Decision Tree

dt_circ = DecisionTreeRegressor(random_state=42)
dt_circ.fit(X_train_circ_scaled, y_train_circ)
pred_circ_dt = dt_circ.predict(X_test_circ_scaled)

# Model 4: Random Forest (chosen)

rf_circ = RandomForestRegressor(n_estimators=100, max_depth=20, random_state=42)
rf_circ.fit(X_train_circ_scaled, y_train_circ)
pred_circ_rf = rf_circ.predict(X_test_circ_scaled)

# Model 5: Gradient Boosting

gb_circ = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
gb_circ.fit(X_train_circ_scaled, y_train_circ)
pred_circ_gb = gb_circ.predict(X_test_circ_scaled)



# PART 2: PREDICTING TOTAL OPERATING INCOME (TARGET)

print("PREDICTING: Total Operating Income")


# Inputs for income model: we REMOVE Total Circulation from inputs (to avoid circular dependency)
X_income = df[features_shared].copy()
y_income = df['Total Operating Income'].copy()

# Split data
X_train_income, X_test_income, y_train_income, y_test_income = train_test_split(
    X_income, y_income, test_size=0.2, random_state=42
)

# Scale the features (separate scaler for income)
scaler_income = StandardScaler()
X_train_income_scaled = scaler_income.fit_transform(X_train_income)
X_test_income_scaled = scaler_income.transform(X_test_income)

# Model 1: Linear Regression (Income)

lr_income = LinearRegression()
lr_income.fit(X_train_income_scaled, y_train_income)
pred_income_lr = lr_income.predict(X_test_income_scaled)

# Model 2: Decision Tree (Income)

dt_income = DecisionTreeRegressor(random_state=42)
dt_income.fit(X_train_income_scaled, y_train_income)
pred_income_dt = dt_income.predict(X_test_income_scaled)

# Model 3: Random Forest (Income)

rf_income = RandomForestRegressor(n_estimators=100, max_depth=20, random_state=42)
rf_income.fit(X_train_income_scaled, y_train_income)
pred_income_rf = rf_income.predict(X_test_income_scaled)

# Model 4: Gradient Boosting (Income)

gb_income = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
gb_income.fit(X_train_income_scaled, y_train_income)
pred_income_gb = gb_income.predict(X_test_income_scaled)



# PART 3: EVALUATE MODELS

def evaluate(y_true, y_pred, model_name, target_name):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    return [model_name, target_name, round(mae, 2), round(rmse, 2), round(r2, 4)]

results = []

# Circulation models
#results.append(evaluate(y_test_circ, pred_circ_dt, "DecisionTree", "Circulation"))
results.append(evaluate(y_test_circ, pred_circ_rf, "RandomForest", "Circulation"))
results.append(evaluate(y_test_circ, pred_circ_gb, "GradientBoosting", "Circulation"))

# Income models
results.append(evaluate(y_test_income, pred_income_lr, "Linear", "Income"))
#results.append(evaluate(y_test_income, pred_income_dt, "DecisionTree", "Income"))
results.append(evaluate(y_test_income, pred_income_rf, "RandomForest", "Income"))
results.append(evaluate(y_test_income, pred_income_gb, "GradientBoosting", "Income"))

df_results = pd.DataFrame(results, columns=["Model", "Target", "MAE", "RMSE", "R²"])
print("\n", df_results)

# PART 4: VISUALIZATIONS - SHOWING WHY MODELS ARE REJECTED OR CHOSEN

# 1. LINEAR REGRESSION - REJECTED (Circulation)

print("LINEAR REGRESSION - REJECTED (Circulation)")


plt.figure(figsize=(7, 5))
plt.scatter(y_test_circ, pred_circ_lr, alpha=0.6, color='orange')
plt.plot([y_test_circ.min(), y_test_circ.max()], 
         [y_test_circ.min(), y_test_circ.max()], 
         'green', linestyle='--', lw=2, label='Perfect Prediction')
plt.xlabel("Actual Circulation")
plt.ylabel("Predicted Circulation")
plt.title("Linear Regression - REJECTED ❌ (Poor Fit)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

lr_r2 = r2_score(y_test_circ, pred_circ_lr)
print(f"R² Score: {lr_r2:.4f}")
print("REASON: Linear Regression assumes a linear relationship... \n")

# 2. POLYNOMIAL REGRESSION 

print("POLYNOMIAL REGRESSION - REJECTED (Circulation)")


plt.figure(figsize=(7, 5))
plt.scatter(y_test_circ, pred_circ_poly, alpha=0.6, color='purple')
plt.plot([y_test_circ.min(), y_test_circ.max()], 
         [y_test_circ.min(), y_test_circ.max()], 
         'green', linestyle='--', lw=2, label='Perfect Prediction')
plt.xlabel("Actual Circulation")
plt.ylabel("Predicted Circulation")
plt.title("Polynomial Regression - REJECTED ❌ (Overfitting)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

poly_r2 = r2_score(y_test_circ, pred_circ_poly)
print(f"R² Score: {poly_r2:.4f}")
print("REASON: Polynomial regression overfits... \n")

# 3. DECISION TREE 

print("DECISION TREE - REJECTED (Circulation)")


residuals_dt = y_test_circ - pred_circ_dt
plt.figure(figsize=(7, 5))
plt.scatter(pred_circ_dt, residuals_dt, alpha=0.6, color='red')
plt.axhline(y=0, color='blue', linestyle='--', lw=2)
plt.xlabel("Predicted Circulation (Decision Tree)")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Decision Tree - REJECTED ❌ (High Variance)")
plt.grid(alpha=0.3)
plt.show()

dt_r2 = r2_score(y_test_circ, pred_circ_dt)
print(f"R² Score: {dt_r2:.4f}")
print("REASON: Decision Tree overfits... \n")

# 4. RANDOM FOREST - CHOSEN (Circulation)

print("RANDOM FOREST - CHOSEN ✓ (Circulation)")

plt.figure(figsize=(7, 5))
plt.scatter(y_test_circ, pred_circ_rf, alpha=0.6, color='blue')
plt.plot([y_test_circ.min(), y_test_circ.max()], 
         [y_test_circ.min(), y_test_circ.max()], 
         'red', linestyle='--', lw=2, label='Perfect Prediction')
plt.xlabel("Actual Circulation")
plt.ylabel("Predicted Circulation")
plt.title("Random Forest - CHOSEN ✓ (Best Performance)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

rf_r2 = r2_score(y_test_circ, pred_circ_rf)
print(f"R² Score (Circulation RF): {rf_r2:.4f}")
print("REASON: Random Forest works well...\n")

# Distribution Comparison - Circulation
print("Distribution Comparison - Circulation")
plt.figure(figsize=(7, 5))
plt.hist(y_test_circ, bins=30, alpha=0.5, color='red', label='Actual', edgecolor='black')
plt.hist(pred_circ_rf, bins=30, alpha=0.5, color='blue', label='Predicted', edgecolor='black')
plt.xlabel("Circulation Values")
plt.ylabel("Frequency")
plt.title("Distribution Match - Random Forest (Circulation)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Now visualizations for Income models (similar)


# 1. LINEAR REGRESSION - Income
print("\n")
print("LINEAR REGRESSION - (Income)")


plt.figure(figsize=(7, 5))
plt.scatter(y_test_income, pred_income_lr, alpha=0.6, color='orange')
plt.plot([y_test_income.min(), y_test_income.max()], 
         [y_test_income.min(), y_test_income.max()], 
         'green', linestyle='--', lw=2, label='Perfect Prediction')
plt.xlabel("Actual Income")
plt.ylabel("Predicted Income")
plt.title("Linear Regression - Income")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Decision Tree - Income residuals
plt.figure(figsize=(7,5))
residuals_income_dt = y_test_income - pred_income_dt
plt.scatter(pred_income_dt, residuals_income_dt, alpha=0.6, color='red')
plt.axhline(y=0, color='blue', linestyle='--', lw=2)
plt.xlabel("Predicted Income (Decision Tree)")
plt.ylabel("Residuals (Actual - Predicted)")
plt.title("Decision Tree - Income Residuals")
plt.grid(alpha=0.3)
plt.show()

# Random Forest Income distribution
plt.figure(figsize=(7,5))
plt.hist(y_test_income, bins=30, alpha=0.5, color='yellow', label='Actual', edgecolor='black')
plt.hist(pred_income_rf, bins=30, alpha=0.5, color='cyan', label='Predicted', edgecolor='black')
plt.xlabel("Income Values")
plt.ylabel("Frequency")
plt.title("Distribution Match - Random Forest (Income)")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

rf_income_r2 = r2_score(y_test_income, pred_income_rf)
print(f"R² Score (Income RF): {rf_income_r2:.4f}")
print("REASON: Random Forest is consistent across both target variables.\n")

# Feature importance (Circulation RF)
print("Feature Importance Analysis - Circulation RF")
plt.figure(figsize=(10, 6))
importances = rf_circ.feature_importances_
indices = np.argsort(importances)[::-1]

plt.bar(range(len(importances)), importances[indices], color='skyblue', edgecolor='black')
plt.xticks(range(len(importances)), [features_shared[i] for i in indices], rotation=45, ha='right')
plt.xlabel("Features")
plt.ylabel("Importance")
plt.title("Feature Importance - Which Factors Matter Most? (Circulation)")
plt.tight_layout()
plt.show()

print("\nTop 3 Most Important Features (Circulation RF):")
for i in range(3):
    idx = indices[i]
    print(f"  {i+1}. {features_shared[idx]}: {importances[idx]:.4f}")
print("\n")

# PART 5: CROSS-VALIDATION (to check if model is stable)

print("CROSS-VALIDATION SCORES (5-Fold)")


cv_scores_circ = cross_val_score(rf_circ, X_train_circ_scaled, y_train_circ, cv=5, scoring='r2')
cv_scores_income = cross_val_score(rf_income, X_train_income_scaled, y_train_income, cv=5, scoring='r2')

print("\nRandom Forest - Circulation:")
print(f"  Mean R²: {cv_scores_circ.mean():.4f}")
print(f"  Std R²:  {cv_scores_circ.std():.4f}")
print(f"  Individual Folds: {[round(s, 4) for s in cv_scores_circ]}")

print("\nRandom Forest - Income:")
print(f"  Mean R²: {cv_scores_income.mean():.4f}")
print(f"  Std R²:  {cv_scores_income.std():.4f}")
print(f"  Individual Folds: {[round(s, 4) for s in cv_scores_income]}")

print("\nREASON: Cross-validation tests the model on 5 different data splits.\n")

# PART 6: SAVE MODELS & SCALERS

print("SAVING MODELS")


pickle.dump(rf_circ, open("models/rf_circulation.pkl", "wb"))
pickle.dump(rf_income, open("models/rf_income.pkl", "wb"))
pickle.dump(scaler_circ, open("models/scaler_circ.pkl", "wb"))
pickle.dump(scaler_income, open("models/scaler_income.pkl", "wb"))


print(f"\nFinal Model R² (Circulation RF): {rf_r2:.4f}")
print(f"Final Model R² (Income RF): {rf_income_r2:.4f}")

```
