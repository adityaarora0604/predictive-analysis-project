```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df=pd.read_csv("Public Libraries.csv")
print(df.head())

print(df.describe())

print(df.dtypes)

df.isna().sum()

df.shape


df=df.drop(columns=['Use of Public Internet Computers'])

num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df.loc[:, col] = df[col].fillna(df[col].median())


df.isnull().sum()

le=LabelEncoder()
df['Library']=le.fit_transform(df['Library'])
df['County']=le.fit_transform(df['County'])

df.to_csv("cleaned_data.csv", index=False)

df_cleaned=pd.read_csv("cleaned_data.csv")

df_cleaned.head()

df_cleaned.isnull().sum()















```
