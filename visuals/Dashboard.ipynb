```python
from IPython.display import Image, display, HTML

def show_title(title):
    display(HTML(f"<h1 style='color:#0066CC'>{title}</h1><hr>"))

def show_subtitle(sub):
    display(HTML(f"<h3 style='color:#9500b3'>{sub}</h3>"))

def show_images(img_list, captions):
    html = "<table style='width:100%;'>"
    html += "<tr>"
    for img, caption in zip(img_list, captions):
        html += f"""
        <td style='text-align:center; padding:10px;'>
            <img src='{img}' width='450'><br>
            <p style='font-size:16px; color:#333'><b>{caption}</b></p>
        </td>
        """
    html += "</tr></table>"
    display(HTML(html))


show_title("Exploratory Data Analysis (EDA)") 
show_images(
    [
        "EDA1.png",
        "EDA2.png",
        "EDA3.png",
        "EDA4.png"
    ],
    [
        "Heatmap — Correlations",
        "Total Operating Income vs Circulation",
        "Population vs Circulation",
        "Top 10 Counties — Circulation"
    ]
)


show_title("Regression Models — Accepted Results")
show_images(
    [
        "Reg1Accepted.png",
        "Reg2Accepted.png"
    ],
    [
        "Random Forest — Circulation (Best Performing)",
        "Random Forest — Operating Income (Best Performing)"
    ]
)



show_title("Regression Models — Rejected Results")
show_images(
    [
        "Reg1Reject.png",
        "Reg2Reject.png",
        "Reg3Reject.png"
    ],
    [
        "Linear Regression — Circulation",
        "Polynomial Regression — Circulation",
        "Decision Tree — Operating Income"
    ]
)



show_title("Classification Results — Model Evaluation")
show_images(
    [
        "Classification DT cm.png"
    ],
    [
        "Confusion Matrix — Decision Tree"
    ]
)



show_title("Clustering (K-Means) Insights")


show_subtitle("Silhouette Scores by K")
show_images(
    [
        "Clustering Scores.png"
    ],
    [
        "Silhouette Evaluation Plot"
    ]
)

show_subtitle("Cluster Distribution (PCA 2D)")
show_images(
    [
        "Count using PCA.png"
    ],
    [
        "Cluster Count & Visualization in 2D-PCA"
    ]
)
```
