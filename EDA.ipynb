```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df=pd.read_csv("cleaned_data.csv")
df.head()


num_cols = df.select_dtypes(include=['int64','float64']).columns

plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
plt.title("Correlation Heatmap")
plt.show()

sns.scatterplot(x='Total Operating Income', y='Total Circulation', data=df)
plt.title("Total Operating Income vs Total Circulation")
plt.show()


sns.scatterplot(x='Population of Service Area', y='Total Circulation', data=df, color="red" )
plt.title("Population vs Total Circulation")
plt.show()

heatmap_fig=plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), cmap='coolwarm', annot=False)
heatmap_fig.savefig("visuals/correlation_heatmap.png")


important_cols = [
    'Population of Service Area',
    'Total Library Visits',
    'Total Circulation',
    'Total Operating Income',
    'Operating Expenditures',
    'Town Tax Appropriation for Library',
    'Library Materials Expenditures'
]

plt.figure(figsize=(10,7))
sns.heatmap(df[important_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Focused Correlation Heatmap")
plt.show()


df.groupby('County')['Total Circulation'].sum().sort_values(ascending=False).head(10).plot(
    kind='bar', figsize=(10,5)
)
plt.title("Top 10 Counties by Total Circulation")
plt.ylabel("Total Circulation")
plt.show()


df.corr()['Total Circulation'].sort_values(ascending=False).head(10)


df.corr()['Total Operating Income'].sort_values(ascending=False).head(10)
```
